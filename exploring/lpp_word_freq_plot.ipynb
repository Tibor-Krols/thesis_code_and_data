```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
```

```python
# Read the text file
path = r'C:\Users\tibor\Documents\thesis\code and data\text'
with open(os.path.join(path,'the_little_prince.txt'), 'r',encoding='utf-8') as file:
# with open('your_text_file.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# Tokenize the text into words
words = word_tokenize(text)
```

```python
# Calculate word frequencies
word_freq = FreqDist(words)
```

```python
total_words = len(words)
zipf_values = [total_words / (i + 1) for i in range(len(word_freq))]
```

```python
vocab_size = len(set(words))
print(f'vocab_size = {vocab_size}')
print(f'total_words = {total_words}')
```

```python
# Get the most common words and their frequencies
common_words = word_freq.most_common(25)  # Change the number as needed
words, frequencies = zip(*common_words)

# Plot the actual word frequencies and Zipf's Law values
plt.figure(figsize=(10, 6))
plt.bar(range(len(words)), frequencies, tick_label=words, color='blue', label='Actual Frequencies')
# plt.plot(zipf_values, color='red', linestyle='dashed', label="Zipf's Law")
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Word Frequency Distribution')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
