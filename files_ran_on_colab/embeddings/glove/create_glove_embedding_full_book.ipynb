```python
import pandas as pd
import numpy as np
from gensim.models import KeyedVectors
import spacy
```

```python
# Load the transformer-based spaCy model
nlp = spacy.load("en_core_web_md")
```

```python
df_word = pd.read_csv('baseline_word_predictions_per_volume_full_text.csv')
```

```python
def embed_volume(text):
  word_embedding = nlp(text).vector
  return word_embedding
```

```python
df_word['GloVe'] = df_word['prediction'].astype(str).apply(embed_volume)


```
