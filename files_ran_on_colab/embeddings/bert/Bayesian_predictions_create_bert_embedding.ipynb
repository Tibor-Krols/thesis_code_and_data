# load data

# bert embeddings

```python
# Load the transformer-based spaCy model
nlp_bert = spacy.load("en_core_web_trf")
def embed_volume_bert(text):
  doc = nlp_bert(text)
  word_emdedding = doc._.trf_data.tensors[-1]
  word_embedding = word_emdedding.reshape(-1)
  return word_embedding

```

# GloVe embeddings


```python
# Load the transformer-based spaCy model
nlp_glove = spacy.load("en_core_web_md")
def embed_volume_glove(text):
  word_embedding = nlp_glove(text).vector
  return word_embedding
```

```python
participants = ['sub-EN057', 'sub-EN058']

for participant in participants:
  # load files
  filename = f'bayes_vol_pred_{participant}_section8_mse_masked_Angular Gyrus_and_more.pkl'
  df_vol = pd.read_pickle(filename)

  # embed preds
  df_vol['pred_GloVe'] = df_vol['pred_text'].astype(str).apply(embed_volume_glove)
  df_vol['pred_BERT'] = df_vol['pred_text'].astype(str).apply(embed_volume_bert)

  # save file
  df_vol.to_pickle(f'bayesian_embeddings_per_volume_{participant}.pkl')


```
